```python
X = pd.read_csv('train_data.csv', dtype=dtype_dict)
train_labels = pd.read_csv('train_labels.csv', dtype={'target': 'int8'})
train_labels.drop(columns=['customer_ID'], inplace=True)
y = np.ravel(train_labels)
del train_labels
gc.collect()
```

```python
numerical = X.drop(columns= categorical).columns.to_list()
numerical.remove('customer_ID')
numerical.remove('S_2')
X_num_agg = X.groupby("customer_ID")[numerical].agg(['mean', 'std', 'min', 'max', 'last'])
X_num_agg.head(20)
```

```python
X_num_agg.columns = ['_'.join(x) for x in X_num_agg.columns]
X_num_agg.reset_index(inplace = True)
```

```python
train_cat_agg = X.groupby("customer_ID")[categorical].agg(['count', 'last', 'nunique'])
del X
gc.collect()
train_cat_agg.columns = ['_'.join(x) for x in train_cat_agg.columns]
train_cat_agg.reset_index(inplace = True)
```

```python
le = LabelEncoder()
for col in train_cat_agg.columns:
    if train_cat_agg[col].dtypes == 'object':
        train_cat_agg[col] = le.fit_transform(train_cat_agg[col])
```

```python
train_cat_agg.drop(['customer_ID'], axis= 1, inplace= True)
X_train = pd.concat([X_num_agg, train_cat_agg], axis= 1)
del X_num_agg
del train_cat_agg
```

```python
X_train.drop(['customer_ID'], axis= 1, inplace= True)
```

```python
# X_train.to_csv('agg_train.csv', index= False )
```

```python
import lightgbm as lgb
d_train = lgb.Dataset(X_train, y)
params = {
    'boosting_type': 'dart',
    'objective': 'binary',
    'learning_rate': 0.04,
    'n_estimators': 1100,
    'num_leaves': 127,
    # 'device': 'gpu'
}
model = lgb.train(params= params, train_set= d_train)
del X_train
del y
gc.collect()
```

```python
import joblib
joblib.dump(model, 'ft_model.joblib')
gc.collect()
```
